```python
"""Implemente um programa que leia dois conjuntos de números reais, armazenando-
os em vetor(lista)es. Após, calcule o produto escalar entre eles. Os conjuntos têm

5 elementos cada. Ao final, imprimir os dois conjuntos e o produto escalar, sendo
que o produto escalar é dado por: x1 ∗ y1 + x2 ∗ y2 + ... + xn ∗ yn."""

conjunto1 = []
conjunto2 = []

for i in range(0,5):
    conjunto1.append(float(input(f"Informe o {i+1}º valor do conjunto 1: ")))
    conjunto2.append(float(input(f"Informe o {i+1}º valor do conjunto 2: ")))

produto_escalar = 0

for i in range(0,5):
    produto_escalar += conjunto1[i] * conjunto2[i]

print(f"O produto escalar entre os conjuntos é: {produto_escalar}")


```
