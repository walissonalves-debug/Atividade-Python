```python
"""Faça um programa que receba do usuário um vetor(lista) com 20 posições. Em
seguida deverá ser impresso o maior e o menor elemento do vetor(lista)."""

vetor = []

for i in range(0,20):
    vetor.append(int(input(f"Informe o {i+1}º valor: ")))

print(f"O maior elemento do vetor é: {max(vetor)}")
print(f"O menor elemento do vetor é: {min(vetor)}")

```
