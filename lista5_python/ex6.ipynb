```python
"""Crie um programa que leia 6 valores inteiros e, em seguida, mostre na tela os
valores lidos na ordem inversa. Utilizar laço de repetição e vetor(lista)."""

valores = []

for i in range(0,6):
    valores.append(int(input(f"Informe o {i+1}º valor: ")))

    valores.reverse()   
    
print(valores)


```
