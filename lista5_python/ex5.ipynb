```python
"""Faça um programa para ler a nota da prova de 15 alunos e armazene num
vetor(lista), calcule e imprima a média geral."""

notas = []
soma = 0

for i in range(0,15):
    notas.append(float(input(f"Informe a nota do {i+1}º aluno: ")))
    soma += notas[i]

media = soma/15 



print(f"A média geral das notas é: {media:.2f}")


```
